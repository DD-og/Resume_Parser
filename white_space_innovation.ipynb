```python
import fitz 
def extract_text_with_color(pdf_path):
    doc = fitz.open(pdf_path)
    text_with_color = []

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        blocks = page.get_text("dict")["blocks"]

        for b in blocks:  # iterate through the text blocks
            for l in b["lines"]:  # iterate through the text lines
                for s in l["spans"]:  # iterate through the text spans
                    font_color = s["color"]
                    text = s["text"]
                    text_with_color.append((text, font_color))
                    
    doc.close()
    
    return text_with_color

pdf_path = "/Users/devmehta/Desktop/dev/level3_resume.pdf"
text_and_colors = extract_text_with_color(pdf_path)
for text, color in text_and_colors:
    print(f"Text: {text}, Color: {color}")
```
